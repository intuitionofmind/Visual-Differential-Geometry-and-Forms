```python
import sympy as sp
from sympy import pprint, simplify, expand, latex
from IPython.display import display
from fractions import Fraction
sp.init_printing(
    use_latex=True,
    pretty_print=True,
    use_unicode=True,
    latex_mode="equation",
    fontsize=48)

from IPython.display import HTML

from sympy import Function, dsolve, Derivative, checkodesol, Eq
```

```python
# 习题5

R, w, q, t = sp.symbols('R,\omega,q,t', real=True)

r = sp.Matrix([R*sp.cos(w*t), R*sp.sin(w*t), q*t])
v = sp.diff(r, t)
a = sp.diff(v, t)
aa = sp.diff(a, t)

display(r)
display(v)
display(a)

kappa = (v.cross(a).norm()) / v.norm()**3
print("曲率")
display(kappa)
display(simplify(kappa))

tau = (v.cross(a)).dot(aa) / (v.cross(a)).norm()**2
print("挠率")
display(tau)
display(simplify(tau))
```
