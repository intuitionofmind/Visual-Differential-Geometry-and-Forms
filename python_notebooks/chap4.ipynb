```python
u = sp.Symbol(r"u", real=True)
v = sp.Symbol(r"v", real=True)

def f(u, v):
    return u**2 + v**2  

g = sp.diff(sp.diff(sp.log(f(u, v)), u), u) 

display(sp.diff(sp.log(f(u, v)), u))
display(simplify(g))
# display(simplify(g)) # second derivative of f with respect to u and v
```

```python
# 球中心投影的度量张量

R, r, t = sp.symbols(r"R,r,\theta", real=True)

x = R*r*sp.cos(t) / sp.sqrt(R**2 + r**2)
y = R*r*sp.sin(t) / sp.sqrt(R**2 + r**2)
z = R**2 / sp.sqrt(R**2 + r**2)

# the vector
v = sp.Matrix([x, y, z])
display("Vector v:", v)

# the tangent vectors
tan_r = simplify(sp.diff(v, r))
tan_t = simplify(sp.diff(v, t))

display(tan_r)
display(tan_t)

# the metric tensor
m00 = simplify(tan_r.dot(tan_r))
m01 = simplify(tan_r.dot(tan_t))
m10 = simplify(tan_t.dot(tan_r))
m11 = simplify(tan_t.dot(tan_t))
m = sp.Matrix([[m00, m01], [m10, m11]])
display("Metric tensor g:", m)
# latex(m)
```

```python
# 球极投影的度量张量

R, r, t = sp.symbols(r"R,r,\theta", real=True)
phi = sp.Symbol(r"\phi", real=True)

# 球坐标参数化
x = 2*R**2*r*sp.cos(t) / (R**2 + r**2)
y = 2*R**2*r*sp.sin(t) / (R**2 + r**2)
z = R*(r**2 - R**2) / (R**2 + r**2)

# the vector
v = sp.Matrix([x, y, z])
display("Vector v:", v)

# the tangent vectors
tan_r = simplify(sp.diff(v, r))
tan_t = simplify(sp.diff(v, t))

display(tan_r)
display(tan_t)

# the metric tensor
m00 = simplify(tan_r.dot(tan_r))
m01 = simplify(tan_r.dot(tan_t))
m10 = simplify(tan_t.dot(tan_r))
m11 = simplify(tan_t.dot(tan_t))
m = sp.Matrix([[m00, m01], [m10, m11]])
display("Metric tensor g:", m)
```

```python
# 习题10、11

# 阿基米德-兰伯特投影

R, x, y = sp.symbols(r"R,x,y", real=True)

A = sp.sqrt(R**2 - y**2) / R
B  = R / sp.sqrt(R**2 - y**2)

# 曲率
K = -1/(A*B)*(sp.diff(sp.diff(A, y) / B, y) + sp.diff(sp.diff(B, x) / A, x))

display("阿基米德-兰伯特投影曲率 K:", simplify(K))

# 圆柱中心投影

A = R / sp.sqrt(R**2 + y**2)
B  = R**2 / (R**2 + y**2)

# 曲率
K = -1/(A*B)*(sp.diff(sp.diff(A, y) / B, y) + sp.diff(sp.diff(B, x) / A, x))

display("圆柱中心投影曲率 K:", simplify(K))
```
